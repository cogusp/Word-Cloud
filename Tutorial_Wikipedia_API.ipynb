- How To Get Page Sections

```python
def print_sections(sections, level=0):
    for s in sections:
        print("%s: %s - %s" % ("*" * (level + 1), s.title, s.text[0:40]))
        print_sections(s.sections, level + 1)
        
print_sections(page_py.sections)
```

- How To Get Page In Other Language

```python
def print_langlinks(page):
    # Get all link to the language of the page
    langlinks = page.langlinks
    
    # Arrange and output language links alphabetically
    for k in sorted(langlinks.keys()):
        v = langlinks[k]
        print("%s: %s - %s: %s" % (k, v.language, v.title, v.fullurl))
        
# Language link output on Wikipedia page
print_langlinks(page_py)

# Output summary of Wikipedia pages for a particular language
page_py_cs = page_py.langlinks['cs']
print("Page - Summary: %s" % page_py_cs.summary[0:5])
```

- How To Get Links To Other Pages

```python
def print_links(page):
    # Get all links on the page
    links = page.links
    
    # Arrange links alphabetically and output them
    for title in sorted(links.keys()):
        # Output the title and link URL for each link
        print("%s: %s" % (title, links[title]))
        
print_links(page_py)
```

- How To Get Page Categories

```python
def print_categories(page):
    # Get all links on the page
    categories = page.categories
    
    # Arrange links alphabetically and output them
    for title in sorted(categories.keys()):
        print("%s: %s" % (title, categories[title]))
        
print("Categories")
print_categories(page_py)
```

- How To Get All Pages From Category

```python
def print_categorymembers(categorymembers, level=0, max_level=1):
    # Repeated processing of category members.
    for c in categorymembers.values():
        # Outputs the title and namespaces of the current member
        print("%s: %s (ns: %d)" % ("*" * (level + 1), c.title, c.ns))
        # If the current member is a category, and the maximum level has not been reached,
        # Output members of that category
        if c.ns == wikipediaapi.Namespace.CATEGORY and level < max_level:
            print_categorymembers(c.categorymembers, level=level+1, max_level=max_level)
            
cat = wiki_wiki.page("Category:Physics")
print("Caterogy members: Category:Physics")
print_categorymembers(cat.categorymembers)
```

- How To See Underlying API Call

```python
import wikipediaapi
import sys

# Set the log level (set to DEBUG level to output debugging information)
wikipediaapi.log.setLevel(level = wikipediaapi.logging.DEBUG)

# Set handler if tou use Python in interactive mode
# Create log handler: output logs using stderr (standard error stream)
out_hdlr = wikipediaapi.logging.StreamHandler(sys.stderr)
# Formatting log messages
out_hdlr.setFormatter(wikipediaapi.logging.Formatter('%(asctime)s %(message)s'))
# Set log level: Set to DEBUG level to output all log messages
out_hdlr.setLevel(wikipediaapi.logging.DEBUG)
# Adding Handlers to Wikipedia Log Objects
wikipediaapi.log.addHandler(out_hdlr)

# Creating Wikipedia Objects
wiki = wikipediaapi.Wikipedia(user_agent='WikipediaAPITutorial (cogusp1@gmail.com)', language='en')

# Get 'Ostrava' page and save it to variable
page_ostrava = wiki.page('Ostrava')
# Summary output on the 'Ostrava' page
print(page_ostrava.summary)
```
